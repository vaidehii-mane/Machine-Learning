EXP - 10 Principal Component Analysis

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
X = np.array([
    [4, 11],
    [8, 4],
    [13, 5],
    [7, 14]
])

scaler = StandardScaler()
X_std = scaler.fit_transform(X)
```

```python
pca = PCA(n_components=1)
Z_reduced = pca.fit_transform(X_std)
```

```python
print("Original Data:\n", X)

print("\nMean of each feature (used in PCA):\n", pca.mean_)

print("\nPrincipal Component (Eigenvector):\n", pca.components_)
print("\nExplained Variance:\n", pca.explained_variance_)
print("\nReduced Data (1D projection):\n", Z_reduced)
```
