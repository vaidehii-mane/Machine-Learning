EXP - 7 Perform K-Nearest Neighbor algorithm by using ready method from sklearn



Import Libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
```

Load Dataset

```python
df=load_iris()
x=pd.DataFrame(df.data,columns=df.feature_names)
y=pd.Series(df.target)
```

Data Preprocessing

```python
print("The first 5 values are: ")
print(x.head())
```

```python
print("\nSummary: ",x.describe())
print("\nMissing Values: ",x.isnull().sum())
print("\nInformation: ",x.info())
print("\nDuplicate values: ",x.duplicated().sum())
print("\nSize of the dataset: ",x.size)
```

Train test split

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

KNN Model

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
y_pred = knn.predict(x_test)

print("Accuracy:( model for knn) ", accuracy_score(y_test, y_pred))
```
